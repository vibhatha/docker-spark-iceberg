```python
import json
import os
```

```python
import ibis
from ibis_substrait.compiler.core import SubstraitCompiler
from google.protobuf import json_format
```

```python
tb = ibis.table([("id", "date"), ("bid", "float"), ("ask", "int32"), ("symbol", "int64")], "t",)
```

```python
expr = tb.select(["id", "bid", "ask", "symbol"])
```

```python
compiler = SubstraitCompiler()
proto = compiler.compile(expr)
json_plan = json_format.MessageToJson(proto)
```

```python
fname_base = '/home/iceberg/substrait/'
substrait_plan_file = fname_base+"read_substrait.json"
```

```python
with open(substrait_plan_file, "w") as f:
    f.write(json_plan)
```

```python
def load_and_print(json_plan:str):
    json_object = json.loads(json_plan)
    json_formatted_str = json.dumps(json_object, indent=2)
    print(json_formatted_str)
```

```python
with open(substrait_plan_file, "rb") as fp:
    substrait_plan_str = fp.read()
```

```python
load_and_print(substrait_plan_str)
```
